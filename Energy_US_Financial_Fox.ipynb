```python
missing_values=energyDF.isnull()
missing_values.head()
```

```python
# Deleting every column where all the entries are missing data
# missing=pd.isnull(energyDF['generation_biomass'])
Energy_Shape=energyDF.shape
count_column=Energy_Shape[1]
count_row=Energy_Shape[0]
print(count_column, count_row)
```

```python
true=0
false=0

for i in range(count_column):
    
    column_name=energyDF.columns[i]
    
    for j in range(count_row):
        if missing_values.iloc[j,i]=="True":
            true=true+1
        else:
            false=false+1
    
    if true==35064:
        energyDF.drop(columns=column_name,inplace=True)
```
